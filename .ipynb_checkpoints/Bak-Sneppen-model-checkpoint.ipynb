## Creating cellular automata model

Steps:
Inititalize percolation grid
Start time loop
For every time iteration:
    Calculate probabilities that neighbours change

```python

def apply_life_rules(cell, neighbors_sum):
    """
    Apply the rules of Conway's Game of Life.
    """
    if cell == 1:
        return 1 if 2 <= neighbors_sum <= 3 else 0
    else:
        return 1 if neighbors_sum == 3 else 0

def generate_2d_ca(initial_state, generations):
    """
    Generate a 2D cellular automaton using Conway's Game of Life rules.
    """
    state = np.array(initial_state)
    ca_history = [state]

    for _ in range(generations):
        new_state = np.zeros_like(state)
        for i in range(state.shape[0]):
            for j in range(state.shape[1]):
                neighbors_sum = np.sum(state[i-1:i+2, j-1:j+2]) - state[i, j]
                new_state[i, j] = apply_life_rules(state[i, j], neighbors_sum)

        state = new_state
        ca_history.append(state.copy())

    return ca_history

def plot_2d_ca(ca_history):
    """
    Plot the 2D cellular automaton using animation.
    """
    fig, ax = plt.subplots()
    img = ax.imshow(ca_history[0], cmap='binary', interpolation='nearest')

    def update(frame):
        img.set_array(ca_history[frame])
        return [img]

    ani = animation.FuncAnimation(fig, update, frames=len(ca_history), interval=200, blit=True)
    plt.title('2D Cellular Automaton (Game of Life)')
    plt.show()

# Set the parameters
initial_state = np.random.choice([0, 1], size=(50, 50), p=[0.9, 0.1])  # Initial random state
generations = 50  # Number of generations

# Generate and plot the 2D cellular automaton
ca_history_2d = generate_2d_ca(initial_state, generations)
plot_2d_ca(ca_history_2d)
```
